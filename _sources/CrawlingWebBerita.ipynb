## crowling
Crawling, atau dikenal juga sebagai web crawling, adalah proses mengunjungi halaman-halaman web secara otomatis menggunakan program komputer yang disebut crawler atau spider.

Tujuan crawling adalah untuk untuk mengidentifikasi suatu laman website agar mesin pencari dapat mengelompokkannya ke kategori tertentu dengan mudah.

dalam mengambil data kita juga pasti tau tentang scraping, perbedaan antara scrapin dan crowling sendiri adalah!

**Scraping**, atau dikenal juga sebagai web scraping, adalah proses mengambil data spesifik dari halaman web dan mengubahnya menjadi format yang berguna (seperti CSV, Excel, atau database). Scraping melibatkan pengambilan konten tertentu dari halaman web (seperti teks, gambar, harga produk, dll.) untuk keperluan analisis, penelitian, atau aplikasi bisnis lainnya.

### mengimport pustaka

```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta
```

- requests digunakan untuk mendapatkan HTML dari web
- BeautifulSoup digunakan untuk memparsing HTML tersebut
- datetime serta timedelta digunakan untuk mengelola dan memanipulasi data waktu

### mengambil konten dari halaman web

```python
url = 'https://www.tempo.co/terpopuler'
response = requests.get(url)
```

- digunakan untuk memparsing HTML dengan BeautifulSoup

### memparsing konten HTML

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

- Digunakan untuk mencari tag tertentu, mendapatkan teks dari elemen, atau mengakses atribut dari tag HTML.

### Mengambil Artikel dengan Kelas 'text-card'

```python
# Temukan semua artikel dengan kelas 'text-card'
articles = soup.find_all('article', class_='text-card')
```

find_all mengembalikan daftar semua elemen yang cocok dengan kriteria yang ditentukan, dalam hal ini semua elemen <article> dengan kelas text-card.

### Mengambil Judul dan Link

```python
article_counter = 1

for article in articles:
    title = article.find('h2', class_='title')
    if title:
        title_text = title.text.strip()
        link = title.find('a')['href']
        print(f"Judul: {title_text}")
        print(f"   Link: {link}")
```

- Mengambil judul:  
    - `article.find('h2', class_='title')` mencari elemen `<h2>` di dalam artikel yang memiliki kelas `title`.
    - Jika elemen ini ditemukan (`if title:`), teks dari elemen tersebut diambil dan dipangkas whitespace di kedua ujungnya (`strip()`), kemudian disimpan dalam variabel `title_text`.

- Mengambil Link:
    - `title.find('a')['href']` mencari elemen `<a>` (link) di dalam elemen `<h2>` yang ditemukan, lalu mengambil atribut `href` (yang berisi URL dari link tersebut).


### Ambil isi artikel

```python
for article in articles:
    content = article.find('p')
    if content:
        print(f"   Isi: {content.text.strip()}")

    # Ambil tanggal publikasi
    date = article.find('h4', class_='date')
    if date:
        date_text = date.text.strip()
```

  - Bagian ini mencari elemen `<p>` di dalam artikel, yang biasanya berisi teks utama atau cuplikan isi artikel.
  - Jika elemen `<p>` ditemukan (`if content:`), teks di dalamnya diambil, whitespace di kedua ujungnya dihilangkan (`strip()`), dan dicetak.


### Ambil tanggal publikasi

```python
for article in articles:
    date = article.find('h4', class_='date')
    if date:
        date_text = date.text.strip()

        # Konversi "X jam lalu" ke waktu spesifik
        if "jam lalu" in date_text:
            hours_ago = int(date_text.split(' ')[0])
            publish_time = datetime.now() - timedelta(hours=hours_ago)
            print(f"   Tanggal: {publish_time.strftime('%d-%m-%Y %H:%M')}")
        else:
            print(f"   Tanggal: {date_text}")
```

  - **Mengambil Tanggal**:
    - Elemen `<h4>` dengan kelas `date` dicari untuk mendapatkan tanggal atau waktu publikasi artikel.
    - Jika elemen tersebut ditemukan (`if date:`), teks di dalamnya diambil dan dipangkas whitespace di sekitarnya (`strip()`).
  - **Konversi "X jam lalu"**:
    - Jika teks tanggal berisi frasa seperti "X jam lalu", teks tersebut diproses untuk mengekstrak jumlah jam (`int(date_text.split(' ')[0])`), dan waktu spesifik dihitung dengan mengurangkan jumlah jam tersebut dari waktu saat ini (`datetime.now()`).
    - Waktu yang dihitung kemudian diformat sebagai `dd-mm-yyyy HH:MM` dan dicetak.
  - **Jika Tanggal dalam Format Lain**:
    - Jika teks tanggal tidak berisi "jam lalu", teks tersebut langsung dicetak.

```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta
import pandas as pd
from IPython.display import display

# Mengambil data dari situs web
url = 'https://www.tempo.co/terpopuler'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Mengambil artikel dari HTML
articles = soup.find_all('article', class_='text-card')

# Inisialisasi list untuk menampung data artikel
data = []

# Looping melalui artikel dan ekstraksi informasi
article_counter = 1  # Inisialisasi penomoran mulai dari 1
for article in articles:
    title = article.find('h2', class_='title')
    title_text = title.text.strip() if title else None

    content = article.find('p')
    content_text = content.text.strip() if content else None

    date = article.find('h4', class_='date')
    if date:
        date_text = date.text.strip()
        # Konversi tanggal jika ada "jam lalu"
        if "jam lalu" in date_text:
            hours_ago = int(date_text.split(' ')[0])
            publish_time = datetime.now() - timedelta(hours=hours_ago)
            date_text = publish_time.strftime('%d-%m-%Y %H:%M')
    else:
        date_text = None

    # Ekstraksi kategori dari URL jika ada
    link = title.find('a')['href'] if title else None
    category_from_url = link.split('.')[0].replace('https://', '').replace('http://', '') if link else None
    category = category_from_url.capitalize() if category_from_url else None

    # Masukkan data ke list
    data.append({
        'Judul Berita': title_text,
        'Isi Berita': content_text,
        'Tanggal Berita': date_text,
        'Kategori Berita': category
    })

    article_counter += 1  # Inkrementasi penomoran

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Mengatur ulang indeks DataFrame agar dimulai dari 1
df = df.reset_index(drop=True)
df.index += 1

# Menggunakan Styler untuk menampilkan tabel dengan border
styled_df = df.style.set_table_styles(
    [{'selector': 'table', 'props': [('border-collapse', 'collapse'), ('width', '100%')]},
     {'selector': 'th, td', 'props': [('border', '1px solid black'), ('padding', '8px'), ('text-align', 'left')]}]
)

# Menampilkan DataFrame dengan gaya border
display(styled_df)
```
